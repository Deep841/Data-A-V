```python
# Import Libraries
import numpy as np
from scipy.spatial import distance
```

```python
#1. Customer Preference Comparison
customer_A = [4, 5, 2, 3, 4]
customer_B = [5, 3, 2, 4, 5]

# Binary preferences
customer_A_binary = [1, 0, 1, 1, 0, 1]
customer_B_binary = [1, 1, 1, 0, 0, 1]
```

```python
# 1. Euclidean Distance

#Formula:
#D_{euclidean}(A,B) = \sqrt{\sum (A_i - B_i)^2}
euclidean_dist = np.linalg.norm(np.array(customer_A) - np.array(customer_B))
print("Euclidean Distance:", euclidean_dist)
```

```python
#2. Manhattan Distance

#Formula:
#D_{manhattan}(A,B) = \sum |A_i - B_i|
manhattan_dist = distance.cityblock(customer_A, customer_B)
print("Manhattan Distance:", manhattan_dist)
```

```python
#3. Cosine Similarity

#Formula:
#\text{Cosine Similarity} = \frac{A \cdot B}{||A|| ||B||}
cosine_sim = 1 - distance.cosine(customer_A, customer_B)
print("Cosine Similarity:", cosine_sim)
```

```python
# 4. Hamming Distance (for Binary Preferences)

#Formula:
#D_{hamming}(A,B) = \frac{\text{Number of different bits}}{\text{Total bits}}
hamming_dist = distance.hamming(customer_A_binary, customer_B_binary) * len(customer_A_binary)
print("Hamming Distance:", hamming_dist)
```

```python
#5. Jaccard Similarity (for Binary Preferences)

#Formula:
#\text{Jaccard Similarity} = \frac{|A \cap B|}{|A \cup B|}
A = np.array(customer_A_binary)
B = np.array(customer_B_binary)
jaccard_sim = np.sum((A & B)) / np.sum((A | B))
print("Jaccard Similarity:", jaccard_sim)
```

```python
#2. Movie Rating Similarity
#Given Data:
user1 = [5, 3, 4, 4, 2]
user2 = [4, 2, 5, 4, 3]
```

```python
#1. Chebyshev Distance

#Formula:
#D_{chebyshev}(A,B) = \max(|A_i - B_i|)
chebyshev_dist = distance.chebyshev(user1, user2)
print("Chebyshev Distance:", chebyshev_dist)
```

```python
#2. Minkowski Distance (p=3)

#Formula:
#D_{minkowski}(A,B) = \left( \sum |A_i - B_i|^p \right)^{1/p}
minkowski_dist = distance.minkowski(user1, user2, p=3)
print("Minkowski Distance (p=3):", minkowski_dist)
```
